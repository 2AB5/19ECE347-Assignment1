```python
import random
import time

def generate_random_bits(num_bits):
    random_bits= [random.choice([0,1]) for _ in range(num_bits)]
    return random_bits
n = 4
num_bits = 2 * n

rst=GPIO(GPIO.get_gpio_pin(0),'out')
cnt=GPIO(GPIO.get_gpio_pin(1),'out')

random_bits = generate_random_bits(num_bits)

array_nx2 = []
for i in range(n):
    row = random_bits[i * 2 : (i+1) * 2]
    array_nx2.append(row)
    
rows = []
print(f"Random {n}x2 array of bits:")
for row in array_nx2:
    rows.append(row)
print("Input bits:\n",rows,"\n")
print("            cnt|rst\n")

for count,reset in rows:
    cnt.write(count)
    rst.write(reset)
    print(f"Input bits:| {count} | {reset} |")
    time.sleep(4)
```
